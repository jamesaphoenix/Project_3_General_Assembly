# Initial EDA + Cleaning

As we will be only looking at the residential housing market, let's exclude all of the commercial and industrial properties.

```python
#Let's create a mask in order to remove the non-residential housing zones from our dataframe.

mask = df['MSZoning'].isin(['A', 'C', 'I'])
```

```python
df = df[~mask]
```

```python
#The following columns were dropped because they had over 1000+ NaN values.
```

```python
df = df.drop(columns=['PoolQC','MiscFeature', 'Alley', 'Fence'])
```

```python
### Cleaning ### 

#In order to preserve data I've decided to fill in any NaN values with either
#the column's mean, None, N/A or a complimentary value to the feature. 

#Replacing Electrical the most regularly occuring variable.
df['Electrical'].fillna('SBrkr', inplace=True)
df['BsmtFinType1'].fillna('Unf', inplace=True)
#Replacing the missing values in the continous data with the mean for that specific column.
df['LotFrontage'].fillna(df['LotFrontage'].mean(), inplace=True)
df['MasVnrArea'].fillna(df['MasVnrArea'].mean(), inplace=True)
df['MasVnrType'].fillna('None', inplace=True)
df['FireplaceQu'].fillna('N/A', inplace=True)
df['BsmtFinType2'].fillna('N/A', inplace=True)
df['BsmtExposure'].fillna('N/A', inplace=True)
df['BsmtCond'].fillna('N/A', inplace=True)
df['BsmtQual'].fillna('N/A', inplace=True)
#Replacing the NaN values with None or N/A to avoid dropping 81 rows.
df.update(df[['GarageCond', 'GarageQual', 'GarageFinish', 
              'GarageYrBlt', 'GarageType']].fillna('N/A'))
```

```python
def correlation_heat_map(df):
    corrs = df.corr()
    corrs = df.corr()[(df.corr() < 1.0) & (df.corr() > 0.5) | (df.corr() < -0.5)].dropna(how='all')

    # Set the default matplotlib figure size:
    fig, ax = plt.subplots(figsize=(20, 9))

    # # Plot the heatmap with seaborn.
    ax = sns.heatmap(corrs, annot=True, cmap='coolwarm')
```

```python
correlation_heat_map(df)
```

```python
#This is an alterative way to get a list of the fixed features.

fixed_features = df.columns.difference(flexible_features)

fixed_df = df.loc[:,fixed_features]
```

```python
#Let's create dummy variables for our fixed features linear regression model.

fixed_df = pd.get_dummies(fixed_df, drop_first=True, columns=None)
```

```python
#Importing all of the models + CV Constructors + KFold + Prediction Scores + Cross Validation
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.metrics import r2_score
#Importing the standardisation module 
from sklearn.preprocessing import StandardScaler
#Import train_test_split
from sklearn.model_selection import train_test_split
```

```python
#The Machine Learning Process
#1. Split Data
#2. Standardise Data
#3. Fit Trained Data
#4. Perform Linear Regression + Cross Validation + Kfold
#5. Perform Regularized Linear Regression (Lasso + Other One + ElasticNet)
#6. Perform Polynomial Linear Regression with the correct model
#7. Plot the residuals against our predicted values
```

```python
y_train = fixed_df[fixed_df['YrSold'] < 2010]['SalePrice']
y_test = fixed_df[fixed_df['YrSold'] == 2010]['SalePrice']
```

```python
#Let's drop the y column before splitting up our X_train and X_test data into two dataframes.
fixed_df.drop(columns=['SalePrice'], axis=1, inplace=True)
```

```python
X_train = fixed_df[fixed_df['YrSold'] < 2010]
```

```python
X_test = fixed_df[fixed_df['YrSold'] == 2010]
```

```python
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

-------------------------------------------------------------------------------------------------------------------------------

# Vanilla Linear Regression Model

```python
vanilla_model = LinearRegression()

vanilla_model.fit(X_train,y_train)
```

```python
vanilla_model.score(X_train,y_train)
```

```python
vanilla_model.score(X_test,y_test)
```

```python
cross_val_score(vanilla_model, X_test, y_test, cv=KFold(n_splits=5))
```

# Calculating the optimal coefficient values for Lasso + Ridge CV + ElasticNetCV

```python
alphas = np.logspace(-5, 5, 100)
```

```python
ridge_test = RidgeCV(alphas ,cv=KFold(n_splits=5))
```

```python
ridge_test.fit(X_train, y_train)
```

```python
ridge_test.alpha_
```

-------------------------------------------------------------------------------------------------------------------------------

```python
lasso_test = LassoCV(alphas=np.logspace(-5, 5, 100), cv=KFold(n_splits=5), max_iter=10000)
```

```python
lasso_test.fit(X_train, y_train)
```

```python
lasso_test.alpha_
```

-------------------------------------------------------------------------------------------------------------------------------

```python
ratios = np.linspace(0.5,1, 1000)
```

```python
elastic_test = ElasticNetCV(l1_ratio = ratios, n_alphas = 500, cv=KFold(n_splits=5), max_iter=10000)
```

```python
elastic_test.fit(X_train, y_train)
```

```python
print(elastic_test.alpha_, round(elastic_test.l1_ratio_, 4))
```

# Testing All Three Models

```python
Ridge_model = Ridge(alpha=475.0810162102793)
Lasso_model = Lasso(alpha=1519.9110829529332)
Elastic_model = ElasticNet(l1_ratio=0.9995, alpha=698.6080476343247)
```

```python
models = [Ridge_model, Lasso_model, Elastic_model]
```

```python
def Model_fitter(models):
    for model in models:
        model.fit(X_train, y_train)
        print("Model Type:", str(model).split('(')[0])
        print("The cross validation scores are", cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5)))
        print("The R2 mean is:", round(cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5)).mean(), 3))
        print("The R2 Standard Deviation is:", round(cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5)).std(), 3) )
        print("---------------------")
```

```python
#Training a model on pre-2010 data, our X_test and y_test are for all of the houses bought within 2010.
Model_fitter(models)
```

## Evaluation of the model - How well does it perform?

After testing three of our models, the ElasticNet (Ridge + Lasso) had the highest R2 mean and lowest R2 standard deviation.

As our R2 is 0.794, approximately 0.8 / 1.0 of our observed values can be explained by our predictor matrix. This model also had the lowest amount of error (RSS) in relation to the TSS. Also because the standard deviation was lowest at 0.033 the ElasticNet model's R2 score has less variance and less bias, making it the model of choice for this specific situation.
