# Testing All Three Models

```python
Ridge_model = Ridge(alpha=599.4842503189409)
Lasso_model = Lasso(alpha=1519.9110829529332)
Elastic_model = ElasticNet(l1_ratio=0.9995, alpha=698.6080476343247)
```

```python
models = [Ridge_model, Lasso_model, Elastic_model]
```

```python
def Model_fitter(models):
    for model in models:
        model.fit(X_train, y_train)
        print("Model Type:", str(model).split('(')[0])
        print("The cross validation scores are", cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5)))
        print("The R2 mean is:", round(cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5)).mean(), 3))
        print("The R2 Standard Deviation is:", round(cross_val_score(model, X_test, y_test, cv=KFold(n_splits=5)).std(), 3) )
        print("---------------------")
```

```python
#Training a model on pre-2010 data, our X_test and y_test are for all of the houses bought within 2010.
Model_fitter(models)
```

## Evaluation of the model - How well does it perform?

After testing three of our models, the ElasticNet (Ridge + Lasso) had the highest R2 mean and lowest R2 standard deviation.

As our R2 is 0.782, approximately 0.8 / 1.0 of our observed values can be explained by our predictor matrix. It also shows us that this model also had the lowest amount of error (RSS) in relation to the TSS. Furthermore the standard deviation was lowest on the ElasticNet model this indicates that it has the lowest variance when comparing multiple R2 scores during cross validation. 

By having a lower standard deviation this means that our model's R2 score is more stable and this particular model has a greater ability to generalize to new test data.

---------------------------------------------------------------------------------------------------------------------

```python
Ridge_model_coefficients = Ridge_model.coef_
Lasso_model_coefficients = Lasso_model.coef_  
ElasticNet_model_coefficients = Elastic_model.coef_  
```

```python
#Let's create a coefficients matrix that combines all of our 3 models into one dataframe.

coefficients_matrix = pd.DataFrame({'Ridge_Coefficients': Ridge_model_coefficients, 
              'Lasso_Coefficients': Lasso_model_coefficients, 
              'ElasticNet_Coefficients': ElasticNet_model_coefficients},
             index=X.columns)

#The ElasticNet model is the most accurate in terms of R2 and standard deviation,
#So the matrix has been sorted by this column.
coefficients_matrix.sort_values(by='ElasticNet_Coefficients', ascending=False).head(10)
```

---------------------------------------------------------------------------------------------------------------------

Let's de-standardize the coefficients so that for every 1 increase of a coefficient feature, <strong> the predicted y_hat price will increase/decrease by the value of that specific coefficient. </strong>

```python
De_standardized_Ridge_Coef = [(Ridge_model_coefficients[i] / fixed_df.iloc[:,i].std()) for i in range(fixed_df.shape[1])]

De_standardized_Lasso_Coef = [(Lasso_model_coefficients[i] / fixed_df.iloc[:,i].std()) for i in range(fixed_df.shape[1])]

De_standardized_ElasticNet_Coef = [(ElasticNet_model_coefficients[i] / fixed_df.iloc[:,i].std()) for i in range(fixed_df.shape[1])]

coefficient_list = [De_standardized_Ridge_Coef, De_standardized_Lasso_Coef, De_standardized_ElasticNet_Coef]

data_dict = {}

for model, coefficient in zip(models, coefficient_list ):
    data_dict['De-standardized_' + str(model).split('(')[0] + '_Coefficients'] = coefficient

de_standardized_coefficients = pd.DataFrame(data_dict, index=X.columns)

de_standardized_coefficients.head(15)
```

##  Two DataFrame's (Train + Test) To Store All Of Our Y-Predictions & Residual Values

```python
#TRAIN_DATA

#Cleaning Up The Data
model_names = [str(model).split('(')[0] for model in models]
predict_dict = {}
y_train.reset_index(drop=True, inplace=True)

#Creating The Columns Dynamically
for model in models:
    predict_dict[str(model).split('(')[0] + ' ' + 'predicted_y_hat']  = list(model.predict(X_train))
train_df = pd.DataFrame(predict_dict)
train_df['Actual_y_hat'] = y_train

#Dynamically Creating The y_hats + Residuals For Every Specific Model Against The Test Set.
for model,name in list(zip(models,model_names)):
    train_df[str(name) + ' ' + 'Residuals'] = train_df[str(name) + ' ' + 'predicted_y_hat'] - train_df['Actual_y_hat']
    
train_df.head(10)
```

```python
#TEST_DATA

#Cleaning Up The Data
model_names = [str(model).split('(')[0] for model in models]
predict_dict = {}
y_test.reset_index(drop=True, inplace=True)

#Creating The Columns Dynamically
for model in models:
    predict_dict[str(model).split('(')[0]] = list(model.predict(X_test))
test_df = pd.DataFrame(predict_dict)
test_df['Actual_y_hat'] = y_test

#Dynamically Creating The y_hats + Residuals For Every Specific Model Against The Test Set.
for model,name in list(zip(models,model_names)):
    test_df[str(name) + ' ' + 'predicted_y_hat'] = model.predict(X_test)
    test_df[str(name) + ' ' + 'Residuals'] = test_df[str(name) + ' ' + 'predicted_y_hat'] - test_df['Actual_y_hat']
    
test_df.head(10)
```

```python
#Stacked Histogram - 3 axes

fix, ax = plt.subplots(figsize=(30,12))
test_df['Ridge Residuals'].hist(alpha=0.3, bins=120, grid=False)
test_df['Lasso Residuals'].hist(alpha=0.3, bins=120, grid=False)
test_df['ElasticNet Residuals'].hist(alpha=0.3, bins=120, grid=False)
plt.title('The Distribution of The Residuals From The Ridge, Lasso and ElasticNet Models', pad=30, fontsize='25')
plt.xlabel('Residuals', fontsize='21', labelpad=30)
plt.ylabel('Counts', fontsize='21', labelpad=30)
plt.legend(['Ridge','Lasso','ElasticNet'], fontsize='18')
plt.show()
```

### 1.  Evaluate the effect in dollars of the renovatable features.

Let's look at all of the houses where we predicted more than the actual value of the property on the first model.

```python
Final_df[Final_df['Lasso predicted_y_hat'] > Final_df['Actual_y_hat']].head(15).sort_values(by='Lasso predicted_y_hat')
```

We can see that for the top houses where our Lasso model produced a greater predicted price, that the Residuals of Residuals are also negative. This means that the amount of error determined by our second model was actually greater than first predicted from the first model.

Therefore the following logic can be applied:
    
1. The lower the Residuals_Of_Residuals, the greater the error there is between the predicted house price and actual house price.
2. The higher our Predicted_Residuals_y_hat is will lead to a more negative Residuals_Of_Residuals number.
3. If the y (residuals from the first model) - yhat (residuals from the second model) == 0 for all of our values, then we would be able to perfectly predict the error from our first model. 
4. Therefore: 
If the Residuals_Of_Residuals value is >= 0, <strong> we can accurately describe the first model's error. </strong> 
If the Residuals_Of_Residuals value is < 0, <strong> we can't accurately describe the first model's error. </strong>

5. This means that the negative coefficients of our second model are the renovatable features that will generate us profit, whilst the positive coefficients of our second model are the features which will likely lead to a decrease in the profitability of the house.
